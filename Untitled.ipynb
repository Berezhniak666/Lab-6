```python
import numpy as np  
import random  

# Функція для генерації масиву  
def generate_array():  
    array = np.random.randint(-64, 65, size=44)  # Генерація масиву з випадковими значеннями  
    num_negatives = random.randint(18, 26)  # Кількість від'ємних значень  
    negative_indices = random.sample(range(44), num_negatives)  # Вибір випадкових індексів для від'ємних значень  

    for i in range(44):  
        if i in negative_indices:  
            array[i] = random.randint(-64, -1)  # Присвоєння від'ємних значень  
        else:  
            array[i] = random.randint(0, 64)  # Присвоєння додатних значень  

    odd_indices = random.sample(range(44), int(0.25 * 44))  # Вибір 25% індексів для парності  
    for i in odd_indices:  
        if array[i] % 2 == 0:  
            array[i] += 1 if array[i] < 64 else -1  # Перетворення на непарні  

    return array  

# Функція для стиснення масиву  
def compress_array(array, threshold):  
    return [x for x in array if x < threshold]  # Залишаємо значення менші за поріг  

# Основна частина коду  
loaded_array = generate_array()  # Виклик функції для генерації масиву  
print("Згенерований масив:", loaded_array)  # Виведення згенерованого масиву  

# Запит користувача про поріг  
threshold_value = int(input("Введіть число для стиснення масиву: "))  
compressed_array = compress_array(loaded_array, threshold_value)  # Виклик функції стиснення  

print("Стиснутий масив:", compressed_array)  # Виведення стиснутого масиву
```

```python
def save_array_to_file(array, filename='array.txt'):  
    np.savetxt(filename, array, fmt='%d')  

# Збереження масиву у файл  
save_array_to_file(array)
```

```python
def load_array_from_file(filename='array.txt'):  
    return np.loadtxt(filename, dtype=int)  

# Зчитування масиву з файлу  
loaded_array = load_array_from_file()  
print("Зчитаний масив:", loaded_array)
```

```python
import matplotlib.pyplot as plt  

def visualize_array(array):  
    odd_elements = array[array % 2 == 1]  
    negative_elements = array[array < 0]  

    plt.figure(figsize=(10, 6))  
    for i, value in enumerate(array):  
        color = 'blue' if value < 0 else 'red' if value % 2 == 1 else 'green'  
        plt.scatter(i, value, color=color, s=25)  

        if value % 2 == 1:  
            plt.text(i, value, str(value), fontsize=9, color='black', ha='center')  

    plt.axhline(y=0, color='black', linestyle='--')  
    plt.title("Візуалізація масиву")  
    plt.xlabel("Індекси")  
    plt.ylabel("Значення")  
    plt.grid()  
    plt.show()  

visualize_array(loaded_array)  
```

```python
def odd_elements_stats(array):  
    odd_elements = array[array % 2 == 1]  
    count = len(odd_elements)  
    average = np.mean(odd_elements) if count > 0 else 0  
    return count, average  

count, average = odd_elements_stats(loaded_array)  
print(f"Кількість непарних елементів: {count}, Середнє: {average}")  
```

```python
def sum_between_negatives(array):  
    negative_indices = np.where(array < 0)[0]  
    
    if len(negative_indices) >= 2:  
        first_neg_index = negative_indices[0]  
        last_neg_index = negative_indices[-1]  
        subarray = array[first_neg_index:last_neg_index + 1]  
        total_sum = np.sum(subarray)  
        count = len(subarray)  
        return total_sum, count  
    return 0, 0  

total_sum, total_count = sum_between_negatives(loaded_array)  
print(f"Сума між першим та останнім від’ємними елементами: {total_sum}, Кількість: {total_count}") 
```

```python
def compress_array(array, threshold):  
    compressed = np.where(array >= threshold, array, 0)  
    return compressed  

# Запит користувача про поріг  
threshold_value = int(input("Введіть число для стиснення масиву: "))  
compressed_array = compress_array(loaded_array, threshold_value)  

print("Стиснутий масив:", compressed_array)  

# Візуалізація  
plt.figure(figsize=(10, 6))  
plt.plot(compressed_array, marker='o')  
plt.axhline(y=threshold_value, color='red', linestyle='--', label='Поріг')  
plt.title("Стиснутий масив")  
plt.xlabel("Індекси")  
plt.ylabel("Значення")  
plt.legend()  
plt.grid()  
plt.show()  
```
